# Demo for flipping_tool
Run a sample BYW web.

```
from webs_toolkit import (
    check_balanced_yamanouchi,
    pair_byw_or_perfect,
    draw_web,
)

word = '121323'
ok, msg, n = check_balanced_yamanouchi(word)
if not ok:
    print('Not a balanced Yamanouchi word:', msg)
else:
    edges = pair_byw_or_perfect(word, n)
    draw_web(word, edges, title_suffix='(demo)')
```
