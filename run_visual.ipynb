# Run Visual Web Toolkit

This notebook lets you input any Balanced Yamanouchi Word (BYW) and visualize the resulting web.

```
from webs_toolkit import (
    check_balanced_yamanouchi,
    pair_byw_or_perfect,
    draw_web,
)

print('Toolkit loaded successfully!')
```

```
word = input('Enter a BYW (e.g., 121323): ').strip()

ok, msg, n = check_balanced_yamanouchi(word)
if not ok:
    print('‚ùå Not a balanced Yamanouchi word:', msg)
else:
    edges = pair_byw_or_perfect(word, n)
    draw_web(word, edges, title_suffix='(interactive)')
```
